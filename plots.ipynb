```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

def add_one_line(folder, algo, N=10):
    t = [ti for ti in pickle.load(open(folder + '/times_' + algo + '.p', 'rb'))]
    v = pickle.load(open(folder + '/values_' + algo + '.p', 'rb'))
    min_v = min(v)
    plt.plot(t, v, linestyle='-', lw = 3 , label=algo)
    return min_v

def plot_logs(dataset='rcv1', asynch_gd=True, synch_gd=True, PIAG=False, DAGA=False, N=10):
    plt.yscale('log')
    plt.grid(True, which='major',linestyle='-',color='0.7')
    plt.grid(True, which='minor',linestyle='--',color='0.7')
    folder = 'logs_' + dataset
    min_v = np.inf
    if asynch_gd:
        min_v = min(min_v, add_one_line(folder, 'asynch_gd'))
    if synch_gd:
        min_v = min(min_v, add_one_line(folder, 'synch_gd'))
    if PIAG:
        min_v = min(min_v, add_one_line(folder, 'asynch_ave'))
    if DAGA:
        min_v = min(min_v, add_one_line(folder, 'daga'))
    plt.xlabel('time', fontsize=18)
    plt.ylabel('accuracy', fontsize=18)
    plt.ylim([min_v, 0.7])
    plt.legend()
```
